```python
import pandas as pd
# exel 스타일 지정 오류 -> warnings로 없애주기
import warnings
warnings.simplefilter("ignore")
```

```python
# 엑셀 파일 불러오기

pop_seoul = pd.read_excel('seoul_people.xlsx')
pop_seoul.head()
```

```python
# 1번째 줄부터 읽도록 옵션 조정, 지정한 컬럼만 읽음

pop_seoul = pd.read_excel('seoul_people.xlsx',
                         header = 1,
                         usecols = 'B, D, E, F, H')

pop_seoul
```

```python
# 컬럼명 변경

pop_seoul.columns = ['구별', '인구수', '한국인', '등록외국인', '고령자']
pop_seoul
```

```python
# 0번 행에 합계 제거

pop_seoul.drop([0], inplace=True)
# inplace = True -> 영구적으로 적용
# default: inplace = False -> 실행시 View형태로 보여짐
pop_seoul
```

```python
# 구별 컬럼을 index로 설정

pop_seoul.set_index('구별', inplace = True)
pop_seoul
```

```python
import matplotlib.pyplot as plt
# 한글 폰트 설정 (한글 깨짐 이슈)
from matplotlib import font_manager, rc

path="ffont.otf"
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font', family=font_name)
```

```python
# pandas 데이터 뒤에 plot 명령, 그래프 종류 - 수평바 그래프

plt.figure()
pop_seoul['인구수'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
# 응용: pop_seoul을 선 그래프로 나타내기

pop_seoul.plot.line()
```

```python
# DataFrame으로 중간값, info, 비교하기
pop_seoul.info
pop_seoul.describe()
```
